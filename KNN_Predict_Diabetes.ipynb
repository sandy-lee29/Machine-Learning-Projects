# KNN - Predict whether a person will have diabetes or not

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
```

```python
dataset = pd.read_csv('KNN_Dataset.csv')
```

```python
print(len(dataset))
dataset.head()
```

```python
dataset.shape
```

```python
# Replace zeroes for key columns that shouldn't be zero
zero_not_accepted = ['Glucose', 'BloodPressure', 'SkinThickness', 'BMI', 'Insulin']
```

```python
# replaced the zeros with means of those columns
for column in zero_not_accepted:
  dataset[column] = dataset[column].replace(0, np.NaN)
  mean = int(dataset[column].mean(skipna=True))
  dataset[column] = dataset[column].replace(np.NaN,mean)
```

```python
# split dataset into training data & test data
X = dataset.iloc[:, 0:8] # column 8 is outcome
y = dataset.iloc[:, 8] # just column 8
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, test_size=0.2) # put aside 20% of data to test later
```

```python
print(len(X_train))
print(len(y_train))
print(len(X_test))
print(len(y_test))
```

```python
# Feature scaling
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

```python
# Define the model: Init K-NN
classifier = KNeighborsClassifier(n_neighbors=11, p=2,metric='euclidean')
```

```python
# Fit Model
classifier.fit(X_train, y_train)
```

```python
# Predict the test set results
y_pred = classifier.predict(X_test)
y_pred
```

```python
# Evaluate Model
cm = confusion_matrix(y_test, y_pred)
print (cm)
print(f1_score(y_test, y_pred))
```

```python
print(accuracy_score(y_test, y_pred))
```
